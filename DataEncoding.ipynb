#Data Encoding

### Nomial/One_Hot_Encoding          (OHE)

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color':['red','blue','green','red','blue','green']
                                                                                        ## create a sample dataframe with a categorical variable
})
```

```python
df
```

```python
### create an instance of one hot encoder
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['color']])
```

```python
### fit the encoder to the dataframe and transform the categorical variable
encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encodded_df=pd.DataFrame(encoder.toarray(),columns=encoder.get_feature_names_out())
```

```python
encodded_df
```

```python
encoder.get_feature_names_out()
```

```python
pd.concat([df,encodded_df],axis=1)
```

## Ordinal encoding

```python
## Orcinal encoding
import pandas as pd
from sklearn.preprocessing   import OrdinalEncoder

df = pd.DataFrame({
    'size':['large','small','medium','large','medium','small']
                                                                                        ## create a sample dataframe with a categorical variable
})
```

```python
df
```

```python
## create an instance of the orcinalEncoder class and fit_transform
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```
