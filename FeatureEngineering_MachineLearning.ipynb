# Missing Values
#### missing values occurs in dataset when some of the informatios is not stored for a variable. There are three mechanisms

### 1 Missing Completely at Random, MCAR:
 Missing completely at random(MCAR) is a type of missing data mechanism in which the probability of a valie being missing is unrelated to both the observed data and the missing data. in other words, if the data is MCAR, the missing values are randomly distributed throughout the dataset, and there is no sustematic reason for whu they are missing.

for example: in a survey avout the prevalence of a certain disease, the missing data might be MCAR if the survey participants with missing values for cetain questions were selected randomlu and their missing responses are not related to their disease status or any other variables measurd in the survey.

### 2 Missing at Random MAR:
 missing at random(mar) is a type of missing data mechanism in which the probability of a value being missing depends only on the observed data, but not on the missing dataitself. In others words, if the data is MAR, the missing values are sustematically related to the observed data, but not to the missing data. Here are a few examples of missing at random:

 income data: Suppose you are collecting income data from a grpup of people, but some participants choose not to report thir income.if the decision to report or not repot income is related to the participant's age or gender, but not to their income level, then the data is missing at random.

 medical data: suppose you are collecting medical data on patients, includding their blood pressure, but some patiens do not reports their blood  pressure are more likely to be youger or have healthier lifestyles, but the missingness is not related to their actual blood pressure values, then the data is missing at random.

 ### 3 Missing data not at random (MNAR):

 it is a type of missing data mechanism where the probability of missing values depends on the value of the missing data itself. In other words, if the data is MNAR, the missingness is not random and is dependent on unobserved or unmeasured factors that are associated with the missing values.

 for example, suppose you are collecting data on the income and job satisfaction of employees in a company. if emloyees who are less satisfied with their jobs are more likely to refuse to report their income, then the data is not missing at random. in this case, the missingness is dependent on job satisfacion, which is not directly observed or measured.

Missing values in machine learning refer to data points where a value is not present or is unavailable for a certain variable in a dataset. They are often represented as blanks, "NaN" (Not a Number), "NA" (Not Available), or other placeholders. Missing values can arise due to various reasons, including:

Data collection errors: Data may not have been collected properly, or errors may have occurred during entry.
Respondent non-response: Participants in surveys or experiments may choose not to answer certain questions.
Technical issues: Data may be lost or corrupted during storage or transmission.
Unavailability of data: Certain information may not be inherently available for certain cases, such as income data for unemployed individuals.
Missing values pose significant challenges in machine learning because most algorithms cannot handle them directly. They can lead to:

Biased results: If missing values are not handled appropriately, they can distort the analysis and lead to incorrect conclusions.
Reduced accuracy: Machine learning models trained on datasets with missing values may have lower predictive accuracy.
Model failure: Some algorithms may not be able to run at all with missing data.
To address missing values, several techniques are commonly employed:

Deletion: Removing rows or columns with missing values. This is simple but can lead to loss of information if a significant amount of data is missing.
Imputation: Filling in missing values with estimated values, such as the mean, median, or mode of the variable, or using more sophisticated techniques like predictive modeling.
Machine learning algorithms that handle missing values: Some algorithms, such as decision trees and random forests, can inherently handle missing values without requiring imputation.
